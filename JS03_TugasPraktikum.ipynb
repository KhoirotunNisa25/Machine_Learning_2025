```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv('/content/wbc(1).csv')
df.head()
```

```python
df.info()
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
# Buang kolom yang tidak berguna
df = df.drop(columns=['id'])
df.info()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.info()
```

```python
df.head(10)
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Cek hasil
print(df[['diagnosis']].head(20), "\n")
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
x = df.drop(columns=['diagnosis']) # fitur
y = df['diagnosis']                # target

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Convert kembali ke DataFrame
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)

x.head()
```

4.
Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

print("Jumlah label data asli:\n", y.value_counts(), "\n")
print("Jumlah label data train:\n", y_train.value_counts(), "\n")
print("Jumlah label data test:\n", y_test.value_counts(), "\n")
```
